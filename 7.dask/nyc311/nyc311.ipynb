### Obtener el Bag del archivo "rows.csv"

```python
import dask.bag as db
from dask.distributed import Client

client = Client(n_workers=4, memory_limit='8GB')
bag_data = db.read_text("rows.csv")
```

### Primeras 5 líneas (datos separados por comas - csv)

```python
bag_data.take(5)
```

### Leer el archivo como un dataframe

```python
import dask.dataframe as dd
df_data_csv = dd.read_csv("rows.csv", parse_dates=['Created Date'], dtype={ "Vehicle Type": "object", 
                                                                            "Taxi Company Borough": "object", 
                                                                            "Bridge Highway Name": "object", 
                                                                            "Bridge Highway Direction": "object", 
                                                                            "Bridge Highway Segment": "object", 
                                                                            "Road Ramp": "object", 
                                                                            "BBL": "float64", 
                                                                            "Facility Type": "object", 
                                                                            "Incident Zip": "float64", 
                                                                            "Taxi Pick Up Location": "object", 
                                                                            "X Coordinate (State Plane)": "float64", 
                                                                            "Y Coordinate (State Plane)": "float64", 
                                                                            "Due Date": "object", 
                                                                            "Incident Zip": "object"})
df_data_csv = df_data_csv.set_index("Unique Key")
```

### Tipos de los datos

```python
df_data_csv.info()
```

### Primeras 5 líneas del dataframe

```python
df_data_csv.head()
```

### ¿Cuántas llamadas hay en total en el archivo?

```python
df_data_csv.count().compute()
```

### ¿Cuál fue la llamada más vieja?

```python
df_data_csv[df_data_csv['Created Date'] == df_data_csv['Created Date'].min()].compute()
```

### ¿Cuál fue la llamada más reciente?

```python
df_data_csv[df_data_csv['Created Date'] == df_data_csv['Created Date'].max()].compute()
```

### ¿Cuáles son las agencias presentes en el dataset?

```python
df_data_csv[["Agency", "Agency Name"]].drop_duplicates().compute()
```

### ¿Cuáles son las diferentes locaciones presentes en el dataset?

```python
df_data_csv[["Location Type"]].drop_duplicates().compute()
```

### ¿Cuáles y cúantas son las diferentes quejas presentes en el dataset?

```python
# df_data_csv.rename(columns={'Unique Key': 'Cantidad', 'Complaint Type': 'Tipo de queja'}).groupby('Tipo de queja')[["Cantidad"]].count().sort_values('Cantidad').compute()
df_data_csv.rename(columns={'Complaint Type': 'Tipo de queja'}).groupby('Tipo de queja')[["Created Date"]].count().sort_values("Created Date").rename(columns={'Created Date': 'Cantidad'}).compute()
```
